```python
df_num=df.isnull().sum()/df.shape[0]*100
df_num
```

```python
# greater then 17 % nan value column
drop_columns=df_num[df_num>17].keys()
drop_columns
```

```python
df_drom_col=df.drop(columns=drop_columns)
```

```python
df_drom_col.isnull().sum()
```

```python
sns.heatmap(df_drom_col.isnull())
```

```python
df2_numr=df_drom_col.select_dtypes(include=['float64','int64'])
df2_numr.head()
```

```python
sns.heatmap(df2_numr.isnull())
```

```python
numeri_var=[var for var in df2_numr.columns if df2_numr[var].isnull().sum()>0] # greater the 0 nan value columns
numeri_var
```

```python
plt.figure(figsize=(16,9))
for i ,var in enumerate(numeri_var):
  plt.subplot(2,2,i+1)
  sns.distplot(df2_numr[var],kde_kws={'color':'red','lw':3})
```

```python
fill_na_mean=df2_numr.fillna(df2_numr.mean())
```

```python
fill_na_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(16,9))
for i , var in enumerate(numeri_var):
  plt.subplot(2,2,i+1)
  sns.distplot(fill_na_mean[var],kde_kws={'color':'red','lw':3})
  sns.distplot(df2_numr[var],kde_kws={'color':'blue','lw':3})
#
```

```python
fill_na_median=df2_numr.fillna(df2_numr.median())
```

```python
fill_na_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(16,9))
for i , var in enumerate(numeri_var):
  plt.subplot(2,2,i+1)
  sns.distplot(fill_na_median[var],bins=20,kde_kws={'color':'red','lw':3})
  sns.distplot(df2_numr[var],bins=25,kde_kws={'color':'blue','lw':3})
```

```python
plt.figure(figsize=(16,9))
for i , var in enumerate(numeri_var):
  plt.subplot(2,2,i+1)
  sns.distplot(fill_na_median[var],bins=20,kde_kws={'color':'red','lw':3})
  sns.distplot(df2_numr[var],bins=25,kde_kws={'color':'blue','lw':3})
  sns.distplot(fill_na_mean[var],bins=30,kde_kws={'color':'y','lw':3})
```
